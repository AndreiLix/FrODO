# Test Loss

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

# Load data from pickle files
PATH = "/home/andrei/Desktop/PROJECT_ELLIS_COMDO/RESULTS_clusters/ANNs/NoMem/test_loss.pkl"
with open(PATH, "rb") as fp:   # Unpickling
    loss_vanillaGD = pickle.load(fp)

PATH = "/home/andrei/Desktop/PROJECT_ELLIS_COMDO/RESULTS_clusters/ANNs/ShortMem/test_loss.pkl"
with open(PATH, "rb") as fp:   # Unpickling
    loss_ShortMem = pickle.load(fp)

PATH = "/home/andrei/Desktop/PROJECT_ELLIS_COMDO/RESULTS_clusters/ANNs/Fractional/test_loss.pkl"
with open(PATH, "rb") as fp:   # Unpickling    
    loss_fractional = pickle.load(fp)
PATH = "/home/andrei/Desktop/PROJECT_ELLIS_COMDO/RESULTS_clusters/ANNs/Fractional/test_loss.pkl"
with open(PATH, "rb") as fp:   # Unpickling    
    test_loss_fractional = pickle.load(fp)

PATH = "/home/andrei/Desktop/PROJECT_ELLIS_COMDO/RESULTS_clusters/ANNs/Adam/test_loss.pkl"
with open(PATH, "rb") as fp:   # Unpickling
    loss_Adam = pickle.load(fp)

PATH = "/home/andrei/Desktop/PROJECT_ELLIS_COMDO/RESULTS_clusters/ANNs/Nesterov/test_loss.pkl"
with open(PATH, "rb") as fp:   # Unpickling
    loss_Nesterov = pickle.load(fp)

# Calculate mean and range across runs for each method
mean_vanillaGD = np.mean(loss_vanillaGD, axis=0)
min_vanillaGD = np.min(loss_vanillaGD, axis=0)
max_vanillaGD = np.max(loss_vanillaGD, axis=0)

mean_ShortMem = np.mean(loss_ShortMem, axis=0)
min_ShortMem = np.min(loss_ShortMem, axis=0)
max_ShortMem = np.max(loss_ShortMem, axis=0)

mean_fractional = np.mean(loss_fractional, axis=0)
min_fractional = np.min(loss_fractional, axis=0)
max_fractional = np.max(loss_fractional, axis=0)

mean_Adam = np.mean(loss_Adam, axis=0)
min_Adam = np.min(loss_Adam, axis=0)
max_Adam = np.max(loss_Adam, axis=0)

mean_Nesterov = np.mean(loss_Nesterov, axis=0)
min_Nesterov = np.min(loss_Nesterov, axis=0)
max_Nesterov = np.max(loss_Nesterov, axis=0)

# Plot
plt.figure(figsize=(10, 6), dpi=150)

# Plot Vanilla Gradient Descent
plt.plot(mean_vanillaGD, label='Vanilla GD', color='green')
plt.fill_between(range(len(mean_vanillaGD)), min_vanillaGD, max_vanillaGD, color='green', alpha=0.1)

# Plot Short Memory
plt.plot(mean_ShortMem, label='Short Memory', color='red')
plt.fill_between(range(len(mean_ShortMem)), min_ShortMem, max_ShortMem, color='red', alpha=0.1)

# Plot Fractional SGD
plt.plot(mean_fractional, label='Fractional SGD', color='blue')
plt.fill_between(range(len(mean_fractional)), min_fractional, max_fractional, color='blue', alpha=0.1)

# Plot Adam
plt.plot(mean_Adam, label='Adam', color='orange')
plt.fill_between(range(len(mean_Adam)), min_Adam, max_Adam, color='orange', alpha=0.1)

# Plot Nesterov
plt.plot(mean_Nesterov, label='Nesterov', color='purple')
plt.fill_between(range(len(mean_Nesterov)), min_Nesterov, max_Nesterov, color='purple', alpha=0.1)


print("len(mean_Nesterov)= ", len(mean_Nesterov))

# plt.xticks(np.linspace(0, 61-1, 61), np.arange(0, 601, 10)) 
# plt.xticks(np.arange(0, len(mean_Nesterov), 20), np.arange(0, len(mean_Nesterov)*10 + 1, 20*10)) 

print("ticks: ", len(np.arange(0, 61, 1.5)))
print("labels: ", len(np.arange(0, 601, 15) ))

# plt.xticks(ticks= np.linspace(0, 61, 30), labels= [ int(x) for x in np.linspace(0, 601, 30) ] ) 


# plt.xticks(ticks= np.arange(0, 61, 3), labels= np.arange(0, 601, 30) ) 
plt.xticks(ticks= np.arange(0, 61, 6), labels= np.arange(0, 601, 60) ) 





# plt.xticks(np.arange(0, 601, 20))


           
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Test-set Loss (mean and range across runs)')
plt.legend()
plt.grid(True)
plt.show()
```
