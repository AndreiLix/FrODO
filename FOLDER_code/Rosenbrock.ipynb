### ${\displaystyle f_{global}(x,y)=(a-x)^{2}+b(y-x^{2})^{2}}$

$f_{1}(x,y)=(a-x)^{2}$

$f_{2}(x,y)=b(y-x^{2})^{2}$

- deafault a = 1
    - controls global minimum = $(a, a^2)$
- default b = 100
    - controls skewness (smaller b, flatter surface) 

```python
import numpy as np
import matplotlib.pyplot as plt

def rosen1(x, a=1.0, b=100.0):

    # Global minimum at x1=a, x2 = any

    x1 = x[0]
    # x2 = x[1]

    return (a - x1)**2


def rosen2(x, a=1.0, b=100.0):

    x1 = x[0]
    x2 = x[1]

    return (b * (x2 - x1**2)**2)



def rosen_global(x, a= 1.0, b=100.0):

    # Global minimum at (a, a**2)

    x1 = x[0]
    x2 = x[1]

    return (a - x1)**2 + (b * (x2 - x1**2)**2) 





# Generate data for the heatmap
x1_values = np.linspace(-10, 10, 100)
x2_values = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1_values, x2_values)
Z = np.zeros_like(X1)

upper_clip = 300000

for i in range(len(x1_values)):
    for j in range(len(x2_values)):
        Z[i, j] = min( rosen_global([X1[i, j], X2[i, j]]), upper_clip)

# Create a heatmap
plt.figure(figsize=(8, 6), dpi= 100)
heatmap = plt.contourf(X1, X2, Z, cmap='viridis', levels=50)

# Add colorbar
plt.colorbar(heatmap, label='value')

plt.scatter(1, 1, label = "global optimum", color = "red")
plt.scatter(-1, 1, label = "private optima", color = "orange")
plt.scatter(3, 9, color = "orange")
plt.scatter(0, 0, label = "initial condition", color = "white")

plt.legend(loc= "lower right")

# Set labels and title
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Rosenbrock Global (f1 + f2) ')

# Show the plot
plt.show()

# start at 0, 0


```
