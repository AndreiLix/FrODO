### ${\displaystyle f_{global}(x,y)=(a-x)^{2}+b(y-x^{2})^{2}}$

$f_{1}(x,y)=(a-x)^{2}$

$f_{2}(x,y)=b(y-x^{2})^{2}$

- deafault a = 1
    - controls global minimum = $(a, a^2)$
- default b = 100
    - controls skewness (smaller b, flatter surface)
