# How do I turn past accumulation of error into a function I can feed to `differint`?

```python
import numpy as np
import differint.differint as df
import matplotlib.pyplot as plt
def f(x):
   return x**2

def f_firstD(x):
   return 2 * x

alpha1 = 0.99
alpha2 = 0.5
alpha3 = 0.1

plt.plot(range(10), [f(x) for x in range(10)], label = "$f$")

DF = df.RL(alpha= alpha1, f_name= f, domain_start= 0, domain_end= 3, num_points= 10)
plt.plot(range(len(DF)), DF, label = f"alpha = {alpha1}")

DF = df.RL(alpha= alpha2, f_name= f, domain_start= 0, domain_end= 3, num_points= 10)
plt.plot(range(len(DF)), DF, label = f"alpha = {alpha2}")

DF = df.RL(alpha= alpha3, f_name= f, domain_start= 0, domain_end= 3, num_points= 10)
plt.plot(range(len(DF)), DF, label = f"alpha = {alpha3}")


plt.plot(range(10), [f_firstD(x) for x in range(10)], label = "alpha = 1")

plt.legend()
plt.show()
```
