## Do barplots + inspect sensitivity to individual hyperparameters

This is especially for memory terms - see if they're redundant by looking if performance drops when beta = 0



```python
import pickle 
from comdo.utils import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("/home/andrei/Desktop/PROJECT_ELLIS_COMDO/FOLDER_code/saved_performance_dict.pkl", "rb") as f:
    performance_dict = pickle.load(f)

print("number simulations: ", len(performance_dict.keys()))

print(performance_dict.items())
```

```python
fractional_steep = {}
fractional_flat = {}

constant_steep = {}
constant_flat = {}

linear_steep = {}
linear_flat = {}

exponential_steep = {}
exponential_flat = {}

for key, value in performance_dict.items():
    if "fractional" in key and (1.0, 0.0) in key:
        fractional_steep[key] = value

    if "fractional" in key and (0.0, 1.0) in key:
        fractional_flat[key] = value

    if "constant" in key and (1.0, 0.0) in key:
        constant_steep[key] = value
    if "constant" in key and (0.0, 1.0) in key:
        constant_flat[key] = value

    if "linear" in key and (1.0, 0.0) in key:
        linear_steep[key] = value
    if "linear" in key and (0.0, 1.0) in key:
        linear_flat[key] = value

    if "exponential" in key and (1.0, 0.0) in key:
        exponential_steep[key] = value
    if "exponential" in key and (0.0, 1.0) in key:
        exponential_flat[key] = value


sorted_fractional_flat = sorted(fractional_flat, key=fractional_flat.get)
sorted_constant_flat = sorted(constant_flat, key=constant_flat.get)
sorted_linear_flat = sorted(linear_flat, key=linear_flat.get)
sorted_exponential_flat = sorted(exponential_flat, key=exponential_flat.get)

sorted_fractional_steep = sorted(fractional_steep, key=fractional_steep.get)
sorted_constant_steep = sorted(constant_steep, key=constant_steep.get)
sorted_linear_steep = sorted(linear_steep, key=linear_steep.get)
sorted_exponential_steep = sorted(exponential_steep, key=exponential_steep.get)
```

## Optimal hyperparameters

```python
print( sorted_linear_flat) #  I expect to see only profile "exponential" and only (0.0, 1.0)
```

```python
print([ performance_dict[key] for key in sorted_linear_steep]) # I expect to see the values sorted in ascending order
```

```python
key_bestHyperparameter_constant_flat = sorted_constant_flat[0]
key_bestHyperparameter_linear_flat = sorted_linear_flat[0]
key_bestHyperparameter_exponential_flat = sorted_exponential_flat[0]
key_bestHyperparameter_fractional_flat = sorted_fractional_flat[0]

key_bestHyperparameter_constant_steep = sorted_constant_steep[0]
key_bestHyperparameter_linear_steep = sorted_linear_steep[0]
key_bestHyperparameter_exponential_steep = sorted_exponential_steep[0]
key_bestHyperparameter_fractional_steep = sorted_fractional_steep[0]


print("Best hyperparameters are:")

print(key_bestHyperparameter_constant_flat )
print(key_bestHyperparameter_linear_flat)
print(key_bestHyperparameter_exponential_flat)
print(key_bestHyperparameter_fractional_flat)
print(key_bestHyperparameter_constant_steep)
print(key_bestHyperparameter_linear_steep)
print(key_bestHyperparameter_exponential_steep)
print(key_bestHyperparameter_fractional_steep)
```

```python
bestHyperparameter_constant_flat = performance_dict[sorted_constant_flat[0]]
bestHyperparameter_linear_flat = performance_dict[sorted_linear_flat[0]]
bestHyperparameter_exponential_flat = performance_dict[sorted_exponential_flat[0]]
bestHyperparameter_fractional_flat = performance_dict[sorted_fractional_flat[0]]

bestHyperparameter_constant_steep = performance_dict[sorted_constant_steep[0]]
bestHyperparameter_linear_steep = performance_dict[sorted_linear_steep[0]]
bestHyperparameter_exponential_steep = performance_dict[sorted_exponential_steep[0]]
bestHyperparameter_fractional_steep = performance_dict[sorted_fractional_steep[0]]
```

```python
print(bestHyperparameter_fractional_steep)
```

```python
steep_data = [bestHyperparameter_constant_steep, bestHyperparameter_linear_steep, bestHyperparameter_exponential_steep, bestHyperparameter_fractional_steep]
flat_data = [bestHyperparameter_constant_flat, bestHyperparameter_linear_flat, bestHyperparameter_exponential_flat, bestHyperparameter_fractional_flat]

# Create bar plot
bar_width = 0.35
index = np.arange(4)

fig, ax = plt.subplots()
bar1 = ax.bar(index , steep_data, bar_width, label='Steep initial gradient', color='orange')
bar2 = ax.bar(index+ bar_width, flat_data, bar_width, label='Flat initial gradient', color='blue')

ax.set_xlabel('Memory Profiles')
ax.set_ylabel('Iterations until convergence')
ax.set_ylim(0, 200)


ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(['Constant', 'Linear', 'Exponential', 'Fractional'])
ax.legend()
ax.set_title("Performance with best hyperparameters by profile")
plt.show()
```
